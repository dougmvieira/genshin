```python
from math import sqrt
from typing import Optional

import ipywidgets as widgets
import pandas as pd
from IPython.display import display

ELEMENTS = ['anemo', 'cryo', 'electro', 'geo', 'hydro', 'pyro', 'physical']
ELEMENT_DAMAGE_BONUSES = [
    "Anemo DMG Bonus",
    "Cryo DMG Bonus",
    "Geo DMG Bonus",
    "Hydro DMG Bonus",
    "Pyro DMG Bonus",
    "Physical DMG Bonus",
]


def optimize_crit_rate_unc(base_atk, flat_atk, n_rolls):
    s = 0.033 * 0.066
    f = flat_atk
    b = 0.04955 * base_atk
    n = n_rolls
    print(s * (b ** 2 * n ** 2 * s - 12 * b ** 2 + 2 * b * f * n * s + f ** 2 * s))
    return (
        b * n * s + f * s + sqrt(
            s * (b ** 2 * n ** 2 * s - 12 * b ** 2 + 2 * b * f * n * s + f ** 2 * s)
        )
    ) / ( 6 * b * s * 0.033)


def optimize_frac_atk_unc(optimal_crit_rate_unc, n_rolls):
    return 0.04955 * (n_rolls - 2 * optimal_crit_rate_unc / 0.033)


def optimize_crit_dmg_unc(optimal_crit_rate_unc, optimal_frac_atk_unc, n_rolls):
    return 0.066 * (n_rolls - optimal_crit_rate_unc / 0.033 - optimal_frac_atk_unc / 0.04955)


def get_base_atk(char_name, weapon_name, char_lvl, weapon_lvl):
    char_base_atk = data['characters'][char_name]['stats'][char_lvl]['atk']
    weapon_base_atk = data['weapons'][weapon_name]['stats'][weapon_lvl]['atk']
    return char_base_atk + weapon_base_atk


def get_atk(char_name, weapon_name, char_lvl, weapon_lvl, sands_stat, atk_rolls, buffs):
    base_atk = get_base_atk(char_name, weapon_name, char_lvl, weapon_lvl)
    flat_atk_bonus = data['artifact_slots']['feather']['atk']
    frac_atk_bonus = 0.04955 * atk_rolls
    if sands_stat == 'ATK':
        frac_atk_bonus += data['artifact_slots']['sands'][sands_stat]
    frac_atk_bonus += buffs.get('ATK (%)', 0)
    total_atk = (1 + frac_atk_bonus) * base_atk + flat_atk_bonus
    return total_atk, frac_atk_bonus, base_atk, flat_atk_bonus


def get_hp(char_name, char_lvl):
    base_hp = data['characters'][char_name]['stats'][char_lvl]['hp']
    flat_hp_bonus = data['artifact_slots']['flower']['hp']
    return base_hp + flat_hp_bonus


def get_def(char_name, char_lvl):
    base_def = data['characters'][char_name]['stats'][char_lvl]['def']
    return base_def


def get_crit_rate(circlet_stat, crit_rate_rolls, artifact_effects, buffs):
    crit_rate = 0.05
    if circlet_stat == 'CRIT Rate':
        crit_rate += data['artifact_slots']['circlet'][circlet_stat]
    crit_rate += artifact_effects.get('CRIT Rate', 0)
    crit_rate += buffs.get('CRIT Rate', 0)
    crit_rate += 0.033 * crit_rate_rolls
    return crit_rate


def get_crit_dmg(char_name, weapon_name, char_lvl, weapon_lvl, circlet_stat, crit_dmg_rolls):
    crit_dmg = 0.5
    if circlet_stat == 'CRIT DMG':
        crit_dmg += data['artifact_slots']['circlet'][circlet_stat]
    crit_dmg += 0.066 * crit_dmg_rolls
    crit_dmg += data['characters'][char_name]['stats'][char_lvl].get('crit_dmg', 0)
    crit_dmg += data['weapons'][weapon_name]['stats'][weapon_lvl].get('crit_dmg', 0)
    return crit_dmg


def get_dmg_bonus(weapon_name, weapon_stacks, goblet_stat, artifact_effects):
    dmg_bonus = {elem: 0. for elem in ELEMENTS}
    for elem, elem_bonus in zip(ELEMENTS, ELEMENT_DAMAGE_BONUSES):
        dmg_bonus[elem] += data['weapons'][weapon_name]['passive'][weapon_stacks].get(elem_bonus, 0)
        if goblet_stat == elem_bonus:
            dmg_bonus[elem] += data['artifact_slots']['goblet'][goblet_stat]
        dmg_bonus[elem] += artifact_effects.get(elem_bonus, 0)
    return dmg_bonus


def compile_artifact_set_effects(first_set_name, second_set_name, enemy_affected_by_cryo, enemy_frozen):
    first_effect = {} if first_set_name == 'None' else data['artifact_sets'][first_set_name]['2-piece']
    if first_set_name == second_set_name:
        if second_set_name == 'None':
            second_effect = {}
        elif second_set_name == 'Blizzard Strayer':
            stacks = enemy_affected_by_cryo + enemy_frozen
            second_effect = data['artifact_sets'][second_set_name]['4-piece'][stacks]
    else:
        second_effect = {}  if second_set_name == 'None' else data['artifact_sets'][second_set_name]['2-piece']
    effects = first_effect.copy()
    for stat, bonus in second_effect.items():
        effects[stat] = effects.get(stat, 0) + bonus
    return effects


def compile_buffs(cryo_resonance, pyro_resonance, is_rosaria_buff, rosaria_crit_rate, enemy_affected_by_cryo):
    buffs = {}
    if cryo_resonance and enemy_affected_by_cryo:
        buffs['CRIT Rate'] = buffs.get('CRIT Rate', 0) + 0.15
    if pyro_resonance:
        buffs['ATK (%)'] = buffs.get('ATK (%)', 0) + 0.25
    if is_rosaria_buff:
        buffs['CRIT Rate'] = buffs.get('CRIT Rate', 0) + 0.15 * min(1, rosaria_crit_rate)
    return buffs


def get_crit_mult(crit_rate, crit_dmg):
    return 1 + crit_rate * crit_dmg


def get_dmg_bonus_mult(char_name, talent_type, dmg_bonus):
    element = data['characters'][char_name]['talents'][talent_type]['element']
    return 1 + dmg_bonus[element]


def get_res_mult(char_name, talent_type, enemy_res):
    element = data['characters'][char_name]['talents'][talent_type]['element']
    res = enemy_res[element]
    if res < 0:
        return 1 - res / 2
    elif res < 0.75:
        return 1 - res
    else:
        return 1 / (4 * res + 1)


def get_def_mult(char_lvl, enemy_lvl):
    return (char_lvl + 100) / (char_lvl + enemy_lvl + 200)


def get_dmg_table(char_name, talent_type, talent_lvl, atk, crit_dmg, crit_mult, dmg_bonus_mult, res_mult, def_mult):
    talent = data['characters'][char_name]['talents'][talent_type]
    talent_scaling = pd.Series(talent['scaling'][talent_lvl - 1])
    return pd.DataFrame(
        {
            'Normal hit': atk * talent_scaling * dmg_bonus_mult * res_mult * def_mult,
            'Critical hit': atk * talent_scaling * dmg_bonus_mult * res_mult * def_mult * (1 + crit_dmg),
            'Average hit': atk * talent_scaling * dmg_bonus_mult * res_mult * def_mult * crit_mult,
        }
    )


def get_enemy_res(enemy_name):
    return dict(
        anemo=data['enemies'][enemy_name]['Anemo RES'],
        cryo=data['enemies'][enemy_name]['Cryo RES'],
        electro=data['enemies'][enemy_name]['Electro RES'],
        geo=data['enemies'][enemy_name]['Geo RES'],
        hydro=data['enemies'][enemy_name]['Hydro RES'],
        pyro=data['enemies'][enemy_name]['Pyro RES'],
        physical=data['enemies'][enemy_name]['Physical RES'],
    )


def build_output_stat_widget(description: Optional[str] = None, width: str = '150px') -> widgets.Text:
    kwargs = dict(disabled=True, layout=widgets.Layout(width=width),)
    if description is not None:
        kwargs['description'] = description
    return widgets.Text(**kwargs)


def build_wrap_box(children: list, **layout_kwargs) -> widgets.Box:
    full_layout_kwargs = dict(
        display='flex',
        flex_flow=f'row wrap',
        align_items='center',
        width='500px',
    )
    full_layout_kwargs.update(layout_kwargs)
    return widgets.Box(
        children,
        layout=widgets.Layout(**full_layout_kwargs),
    )


def update_artifact_rolls(crit_rate, crit_dmg, crit_mult, base_atk, total_atk):
    total_rolls = (
        crit_rate_rolls_widget.value
        + crit_dmg_rolls_widget.value
        + atk_rolls_widget.value
    )
    crit_rate_marginal = 0.033 * crit_dmg / crit_mult
    crit_dmg_marginal = 0.066 * crit_rate / crit_mult
    atk_marginal = 0.04955 * base_atk / total_atk
    total_rolls_widget.value = total_rolls
    crit_rate_marginal_widget.value = f"{crit_rate_marginal:.2%}"
    crit_dmg_marginal_widget.value = f"{crit_dmg_marginal:.2%}"
    atk_marginal_widget.value = f"{atk_marginal:.2%}"


def update_char_stats(artifact_effects, buffs):
    max_hp = get_hp(
        char_name_widget.value,
        char_lvl_widget.value,
    )
    total_atk, _, base_atk, _ = get_atk(
        char_name_widget.value,
        weapon_name_widget.value,
        char_lvl_widget.value,
        weapon_lvl_widget.value,
        sands_stat_widget.value,
        atk_rolls_widget.value,
        buffs,
    )
    total_def = get_def(
        char_name_widget.value,
        char_lvl_widget.value,
    )
    crit_rate = get_crit_rate(
        circlet_stat_widget.value,
        crit_rate_rolls_widget.value,
        artifact_effects,
        buffs,
    )
    crit_dmg = get_crit_dmg(
        char_name_widget.value,
        weapon_name_widget.value,
        char_lvl_widget.value,
        weapon_lvl_widget.value,
        circlet_stat_widget.value,
        crit_dmg_rolls_widget.value,
    )
    max_hp_widget.value = f"{max_hp:,}"
    atk_widget.value = f"{total_atk:,}"
    def_widget.value = f"{total_def:,}"
    crit_rate_widget.value = f"{crit_rate:.1%}"
    crit_dmg_widget.value = f"{crit_dmg:.1%}"
    return total_atk, crit_rate, crit_dmg, base_atk


def update_dmg_bonus(artifact_effects):
    dmg_bonus = get_dmg_bonus(
        weapon_name_widget.value,
        weapon_stacks_widget.index,
        goblet_stat_widget.value,
        artifact_effects,
    )
    for elem in ELEMENTS:
        dmg_bonus_widgets[elem].value = f"{dmg_bonus[elem]:.1%}"
    return dmg_bonus


def update_enemy_res():
    enemy_res = get_enemy_res(enemy_name_widget.value)
    for elem in ELEMENTS:
        enemy_res_widgets[elem].value = f"{enemy_res[elem]:.1%}"
    return enemy_res


def update_dmg_mult(crit_rate, crit_dmg, dmg_bonus, enemy_res):
    char_lvl, _ = map(int, char_lvl_widget.value.split('/'))

    crit_mult = get_crit_mult(crit_rate, crit_dmg)
    dmg_bonus_mult = get_dmg_bonus_mult(char_name_widget.value, talent_type_widget.value, dmg_bonus)
    res_mult = get_res_mult(char_name_widget.value, talent_type_widget.value, enemy_res)
    def_mult = get_def_mult(char_lvl, enemy_lvl_widget.value)

    crit_mult_widget.value = f"{crit_mult:.1%}"
    dmg_bonus_mult_widget.value = f"{dmg_bonus_mult:.1%}"
    res_mult_widget.value = f"{res_mult:.1%}"
    def_mult_widget.value = f"{def_mult:.1%}"
    return crit_mult, dmg_bonus_mult, res_mult, def_mult


def update_dmg_table(atk, crit_dmg, crit_mult, dmg_bonus_mult, res_mult, def_mult):
    dmg_table = get_dmg_table(
        char_name_widget.value,
        talent_type_widget.value,
        talent_lvl_widget.value,
        atk,
        crit_dmg,
        crit_mult,
        dmg_bonus_mult,
        res_mult, 
        def_mult,
    )
    dmg_table_widget.clear_output()
    with dmg_table_widget:
        display(dmg_table.applymap("{:,.0f}".format))


def update_all(_):
    artifact_effects = compile_artifact_set_effects(
        first_set_widget.value,
        second_set_widget.value,
        enemy_status_widgets['cryo'].value,
        enemy_status_widgets['frozen'].value,
    )
    buffs = compile_buffs(
        resonance_widgets['cryo'].value,
        resonance_widgets['pyro'].value,
        rosaria_buff_widget.value,
        rosaria_crit_rate_widget.value,
        enemy_status_widgets['cryo'].value,
    )
    total_atk, crit_rate, crit_dmg, base_atk = update_char_stats(
        artifact_effects, buffs
    )
    dmg_bonus = update_dmg_bonus(artifact_effects)
    enemy_res = update_enemy_res()
    crit_mult, dmg_bonus_mult, res_mult, def_mult = update_dmg_mult(
        crit_rate, crit_dmg, dmg_bonus, enemy_res
    )
    update_artifact_rolls(crit_rate, crit_dmg, crit_mult, base_atk, total_atk)
    update_dmg_table(total_atk, crit_dmg, crit_mult, dmg_bonus_mult, res_mult, def_mult)


def optimize(_):
    artifact_effects = compile_artifact_set_effects(
        first_set_widget.value,
        second_set_widget.value,
        enemy_status_widgets['cryo'].value,
        enemy_status_widgets['frozen'].value,
    )
    buffs = compile_buffs(
        resonance_widgets['cryo'].value,
        resonance_widgets['pyro'].value,
        rosaria_buff_widget.value,
        rosaria_crit_rate_widget.value,
        enemy_status_widgets['cryo'].value,
    )
    _, frac_atk, base_atk, flat_atk = get_atk(
        char_name_widget.value,
        weapon_name_widget.value,
        char_lvl_widget.value,
        weapon_lvl_widget.value,
        sands_stat_widget.value,
        atk_rolls_widget.value,
        buffs,
    )
    crit_rate = get_crit_rate(
        circlet_stat_widget.value,
        crit_rate_rolls_widget.value,
        artifact_effects,
        buffs,
    )
    crit_dmg = get_crit_dmg(
        char_name_widget.value,
        weapon_name_widget.value,
        char_lvl_widget.value,
        weapon_lvl_widget.value,
        circlet_stat_widget.value,
        crit_dmg_rolls_widget.value,
    )
    n_rolls = frac_atk / 0.04955 + crit_rate / 0.033 + crit_dmg / 0.066
    optimal_crit_rate = optimize_crit_rate_unc(base_atk, flat_atk, n_rolls)
    optimal_frac_atk = optimize_frac_atk_unc(optimal_crit_rate, n_rolls)
    optimal_crit_dmg = optimize_crit_dmg_unc(optimal_crit_rate, optimal_frac_atk, n_rolls)
    print(n_rolls, optimal_crit_rate, optimal_frac_atk, optimal_crit_dmg)
    
    optimal_crit_rate_rolls = (optimal_crit_rate - crit_rate) / 0.033 + crit_rate_rolls_widget.value
    optimal_crit_dmg_rolls = (optimal_crit_dmg - crit_dmg) / 0.066 + crit_dmg_rolls_widget.value
    optimal_atk_rolls = (optimal_frac_atk - frac_atk) / 0.04955 + atk_rolls_widget.value

    crit_rate_rolls_widget.value = int(optimal_crit_rate_rolls)
    crit_dmg_rolls_widget.value = int(optimal_crit_dmg_rolls)
    atk_rolls_widget.value = int(optimal_atk_rolls)
    update_all(None)


def switch_tabs(_):
    if tabs.value == 'Home':
        canvas.children = home_widgets
    elif tabs.value == 'Artifacts':
        canvas.children = artifacts_widgets
    elif tabs.value == 'Buffs':
        canvas.children = buffs_widgets


html_style = widgets.HTML(
    """<style>
        .box_style{
            background-image: url('ayaka.png');
            background-color: rgba(255,255,255,0.6);
            background-blend-mode: lighten;
        }
    </style>"""
)

char_name_widget = widgets.Dropdown(
    options=list(data['characters']),
    value='Kamisato Ayaka',
    description='Character:',
    layout=widgets.Layout(width='250px'),
)
char_name_widget.observe(update_all)
char_lvl_widget = widgets.Dropdown(
    options=list(data['characters'][char_name_widget.value]['stats']),
    value='90/90',
    layout=widgets.Layout(width='70px'),
)
char_lvl_widget.observe(update_all)
char_box = widgets.HBox(
    [
        char_name_widget,
        char_lvl_widget,
    ],
)

weapon_name_widget = widgets.Dropdown(
    options=list(data['weapons']),
    description='Weapon:',
    value='Mistsplitter Reforged',
    layout=widgets.Layout(width='250px'),
)
weapon_name_widget.observe(update_all)
weapon_lvl_widget = widgets.Dropdown(
    options=list(data['weapons'][weapon_name_widget.value]['stats']),
    value='90/90',
    layout=widgets.Layout(width='70px'),
)
weapon_lvl_widget.observe(update_all)
weapon_stacks_widget = widgets.Dropdown(
    options=[f'{i} stacks' for i in range(4)],
    value='0 stacks',
    layout=widgets.Layout(width='85px'),
)
weapon_stacks_widget.observe(update_all)
weapon_box = widgets.HBox(
    [
        weapon_name_widget,
        weapon_lvl_widget,
        weapon_stacks_widget,
    ],
)

talent_type_widget = widgets.Dropdown(
    options=['Normal attack', 'Elemental skill', 'Elemental burst'],
    value='Normal attack',
    description='Talent:',
    layout=widgets.Layout(width='250px'),
)
talent_type_widget.observe(update_all)
talent_lvl_widget = widgets.BoundedIntText(
    value=10,
    min=0,
    max=13,
    step=1,
    layout=widgets.Layout(width='72px'),
)
talent_lvl_widget.observe(update_all)
talent_box = widgets.HBox(
    [
        talent_type_widget,
        talent_lvl_widget,
    ],
)

sands_stat_widget = widgets.Dropdown(
    options=list(data['artifact_slots']['sands']),
    value='ATK',
    layout=widgets.Layout(width='145px'),
)
sands_stat_widget.observe(update_all)
sands_stat_box = widgets.VBox([widgets.Label('Sands of Eon:'), sands_stat_widget])
goblet_stat_widget = widgets.Dropdown(
    options=list(data['artifact_slots']['goblet']),
    value='Cryo DMG Bonus',
    layout=widgets.Layout(width='160px'),
)
goblet_stat_widget.observe(update_all)
goblet_stat_box = widgets.VBox([widgets.Label('Goblet of Eonothem:'), goblet_stat_widget])
circlet_stat_widget = widgets.Dropdown(
    options=list(data['artifact_slots']['circlet']),
    value='CRIT DMG',
    layout=widgets.Layout(width='145px'),
)
circlet_stat_widget.observe(update_all)
circlet_stat_box = widgets.VBox([widgets.Label('Circlet of Logos:'), circlet_stat_widget])

artifact_main_box = widgets.HBox(
    [
        sands_stat_box,
        goblet_stat_box,
        circlet_stat_box,
    ],
)

crit_rate_rolls_widget = widgets.FloatSlider(
    value=0,
    min=0,
    max=50.31,
    step=0.118,
    description="CRIT Rate:",
    layout=widgets.Layout(width='auto'),
)
crit_rate_rolls_widget.observe(update_all)
crit_dmg_rolls_widget = widgets.FloatSlider(
    value=0,
    min=0,
    max=50.31,
    step=0.118,
    description="CRIT DMG:",
    layout=widgets.Layout(width='auto'),
)
crit_dmg_rolls_widget.observe(update_all)
atk_rolls_widget = widgets.FloatSlider(
    value=0,
    min=0,
    max=50.31,
    step=0.118,
    description="ATK (%):",
    layout=widgets.Layout(width='auto'),
)
atk_rolls_widget.observe(update_all)
total_rolls_widget = widgets.FloatSlider(
    value=0,
    min=0,
    max=50.31,
    step=0.118,
    description="Total:",
    layout=widgets.Layout(width='auto'),
    disable=True,
)

crit_rate_marginal_widget = build_output_stat_widget(width='auto')
crit_dmg_marginal_widget = build_output_stat_widget(width='auto')
atk_marginal_widget = build_output_stat_widget(width='auto')
optimize_widget = widgets.Button(
    description='Optimize',
    layout=widgets.Layout(width='auto'),
)
optimize_widget.on_click(optimize)

artifact_rolls_box = widgets.GridBox(
    [
        widgets.Label('Artifact substat rolls'),
        widgets.Label('Marginal utility'),
        crit_rate_rolls_widget,
        crit_rate_marginal_widget,
        crit_dmg_rolls_widget,
        crit_dmg_marginal_widget,
        atk_rolls_widget,
        atk_marginal_widget,
        total_rolls_widget,
        optimize_widget,
    ],
    layout=widgets.Layout(
        grid_template_columns='350px 100px',
        grid_template_rows='1fr 1fr 1fr 1fr 1fr',
    ),
)

max_hp_widget = build_output_stat_widget("Max HP:")
atk_widget = build_output_stat_widget("ATK:")
def_widget = build_output_stat_widget("DEF:")
crit_rate_widget = build_output_stat_widget("Crit Rate:")
crit_dmg_widget = build_output_stat_widget("Crit DMG:")

stats_widget_box = build_wrap_box(
    [
        max_hp_widget,
        atk_widget,
        def_widget,
        crit_rate_widget,
        crit_dmg_widget,
    ],
)
stats_box = widgets.VBox(
    [widgets.Label('Character stats'), stats_widget_box]
)

dmg_bonus_widgets = dict(
    anemo=build_output_stat_widget("Anemo:"),
    cryo=build_output_stat_widget("Cryo:"),
    electro=build_output_stat_widget("Electro:"),
    geo=build_output_stat_widget("Geo:"),
    hydro=build_output_stat_widget("Hydro:"),
    pyro=build_output_stat_widget("Pyro:"),
    physical=build_output_stat_widget("Physical:"),
)
dmg_bonus_widget_box = build_wrap_box(list(dmg_bonus_widgets.values()))
dmg_bonus_box = widgets.VBox(
    [widgets.Label('DMG Bonus'), dmg_bonus_widget_box]
)

dmg_bonus_mult_widget = build_output_stat_widget("DMG Bonus:")
def_mult_widget = build_output_stat_widget("Enemy DEF:")
res_mult_widget = build_output_stat_widget("Enemy Res:")
crit_mult_widget = build_output_stat_widget("Crit:")
dmg_mult_widget_box = build_wrap_box(
    [
        dmg_bonus_mult_widget,
        def_mult_widget,
        res_mult_widget,
        crit_mult_widget,
    ]
)
dmg_mult_box = widgets.VBox([widgets.Label('Damage multipliers'), dmg_mult_widget_box])

enemy_name_widget = widgets.Dropdown(
    options=list(data['enemies']),
    value='Hilichurl',
    description='Enemy:',
    layout=widgets.Layout(width='250px'),
)
enemy_name_widget.observe(update_all)
enemy_lvl_widget = widgets.IntText(
    value=100,
    layout=widgets.Layout(width='72px')
)
enemy_lvl_widget.observe(update_all)
enemy_widget = widgets.HBox([enemy_name_widget, enemy_lvl_widget])

enemy_status_widgets = dict(
    cryo=widgets.ToggleButton(description="Affected by Cryo", layout=widgets.Layout(width='150px')),
    electro=widgets.ToggleButton(description='Affected by Electro', layout=widgets.Layout(width='150px')),
    hydro=widgets.ToggleButton(description='Affected by Hydro', layout=widgets.Layout(width='150px')),
    pyro=widgets.ToggleButton(description='Affected by Pyro', layout=widgets.Layout(width='150px')),
    frozen=widgets.ToggleButton(description='Frozen', layout=widgets.Layout(width='150px')),
)
for widget in enemy_status_widgets.values():
    widget.observe(update_all)
enemy_status_widget_box = build_wrap_box(list(enemy_status_widgets.values()))
enemy_status_box = widgets.VBox(
    [widgets.Label('Enemy status'), enemy_status_widget_box]
)

enemy_res_widgets = dict(
    anemo=build_output_stat_widget("Anemo:"),
    cryo=build_output_stat_widget("Cryo:"),
    electro=build_output_stat_widget("Electro:"),
    geo=build_output_stat_widget("Geo:"),
    hydro=build_output_stat_widget("Hydro:"),
    pyro=build_output_stat_widget("Pyro:"),
    physical=build_output_stat_widget("Physical:"),
)
enemy_res_widget_box = build_wrap_box(list(enemy_res_widgets.values()))
enemy_res_box = widgets.VBox(
    [widgets.Label('Enemy resistance'), enemy_res_widget_box]
)

dmg_table_widget = widgets.Output(
    layout=widgets.Layout(
        width='500px',
        height='150px',
        align_items='center',
    )
)

resonance_widgets = dict(
    cryo=widgets.ToggleButton(description='Cryo', layout=widgets.Layout(width='100px')),
    pyro=widgets.ToggleButton(description='Pyro', layout=widgets.Layout(width='100px')),
)
for widget in resonance_widgets.values():
    widget.observe(update_all)
resonance_box = widgets.VBox(
    [widgets.Label('Elemental resonance'), widgets.HBox(list(resonance_widgets.values()))]
)

rosaria_buff_widget = widgets.ToggleButton(description="4th Ascension")
rosaria_buff_widget.observe(update_all)
rosaria_crit_rate_widget = widgets.FloatText(
    description="CRIT Rate:", value=100., min=0., layout=widgets.Layout(width='200px')
)
rosaria_crit_rate_widget.observe(update_all)
rosaria_buff_box = widgets.VBox([widgets.Label('Rosaria'), widgets.HBox([rosaria_buff_widget, rosaria_crit_rate_widget])])

first_set_widget = widgets.Dropdown(
    options=['None', *data['artifact_sets']],
    value='None',
    layout=widgets.Layout(width='150px'),
)
first_set_widget.observe(update_all)
second_set_widget = widgets.Dropdown(
    options=['None', *data['artifact_sets']],
    value='None',
    layout=widgets.Layout(width='150px'),
)
second_set_widget.observe(update_all)
artifact_sets_box = widgets.HBox(
    [
        widgets.Label('Artifact sets:'),
        first_set_widget,
        second_set_widget,
    ]
)

home_widgets = [
    char_box,
    weapon_box,
    talent_box,
    enemy_widget,
    enemy_status_box,
    enemy_res_box,
    stats_box,
    dmg_bonus_box,
    dmg_mult_box,
    dmg_table_widget,
]
artifacts_widgets = [
    artifact_main_box,
    artifact_sets_box,
    artifact_rolls_box,
    stats_box,
    dmg_bonus_box,
]
buffs_widgets = [
    resonance_box,
    rosaria_buff_box,
    stats_box,
    dmg_bonus_box,
]
canvas = build_wrap_box(
    home_widgets,
    flex_flow='column wrap',
    width='1000px', height='500px'
)
canvas.add_class('box_style')

tabs = widgets.ToggleButtons(
    options=["Home", "Artifacts", "Buffs"],
    value="Home",
)
tabs.observe(switch_tabs)

app_widget = widgets.VBox(
    [tabs, canvas, html_style],
)
update_all(None)
app_widget
```
